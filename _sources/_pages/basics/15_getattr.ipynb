# Chapter 15. getattr()

getattr()는 파이썬에서 "문자열 이름으로" 객체의 속성(attribute)에 접근하는 함수이다. 즉, `obj.name`을 **동적으로 접근**할 수 있게 해주는 도구이다. 

## 사용법 

getattr(object, "attribute_name", default_value)

* object: 속성을 가진 객체 
* "attribute_name": '문자열'로 된 속성 이름 
* default_value (optional): 속성이 없을 때 반환값 (/)

## 예제 

```python
class Person:
    name = "Jessi"

p = Person()

print(p.name) # 일반 접근 
print(getattr(p, "name")) # getattr 접근 
print(f"return type of getattr() function: {type(getattr(p, 'name'))}") # getattr 접근 

# 없는 속성을 접근하는 경우, default_value 를 넣어주면 된다. 
print(getattr(p, "age", 0))   # 0 반환 (안전)

# print(getattr(p, "age"))    # AttributeError 발생
```

## 실사용예제

```python
# "python" -> "PYTHON" -> LANGUAGE.PYTHON 
from enum import Enum 

class Language(Enum):
    PYTHON = "python"
    JAVA = "java"
    C = "c"

def process_language(lang: Language):
    if lang == Language.PYTHON:
        print("Current lang is python")
    elif lang == Language.JAVA:
        print("Current lang is java")
    elif lang == Language.C:
        print("Current lang is c")

process_language(Language.C)
```

```python
# "PYTHON" 속성 이름 (string)
# Langugae 객체 
# getattr(Language, "PYTHON") = 그 속성 가져오기 

lang = getattr(Language, "PYTHON") # 내부적으로 Language.PYTHON으로 동작 

print(lang)         # Language.PYTHON
print(type(lang))   # <enum 'Language'>
print(lang.value)   # "python"
```
