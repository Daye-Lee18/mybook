# Chapter 7. Visualization Practice — 파라미터 실험

파라미터 하나를 바꾸면 그래프가 어떻게 달라지는지 직접 비교합니다.  
각 섹션은 **기준(Base)** 그래프와 **파라미터를 바꾼** 그래프를 나란히 보여줍니다.

| 섹션 | 실험 파라미터 |
|------|-------------|
| 히스토그램 | `bins`, `kde`, `stat`, `color` |
| 박스플롯 / 바이올린 | `hue`, `inner`, `showfliers`, `width` |
| 산점도 | `alpha`, `size`, `hue`, `style` |
| 막대그래프 | `errorbar`, `estimator`, `order`, `palette` |
| 라인차트 | `marker`, `linestyle`, `errorbar` |

## 데이터 준비

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (Colab에서 폰트 설치 후 주석 해제)
# plt.rc('font', family='NanumBarunGothic')

df = sns.load_dataset('mpg')
penguins = sns.load_dataset('penguins').dropna()
df.head()
```

## 히스토그램 파라미터 실험

`sns.histplot()` / `plt.hist()`

### `bins` — 구간 수

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, b in zip(axes, [5, 20, 60]):
    sns.histplot(data=df, x='mpg', bins=b, ax=ax)
    ax.set_title(f'bins={b}')
    ax.set_xlabel('mpg')

plt.suptitle('bins 값에 따른 히스토그램 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `kde` — 밀도 곡선 오버레이 (True / False)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(data=df, x='mpg', bins=20, kde=False, ax=axes[0])
axes[0].set_title('kde=False (기본값)')

sns.histplot(data=df, x='mpg', bins=20, kde=True, ax=axes[1])
axes[1].set_title('kde=True')

plt.tight_layout()
plt.show()
```

### `stat` — Y축 기준값

| 값 | 설명 |
|----|------|
| `'count'` | 빈도수 (기본값) |
| `'frequency'` | 구간 너비로 나눈 빈도 |
| `'probability'` | 전체 합이 1이 되도록 정규화 |
| `'density'` | 면적 합이 1 (KDE와 스케일 맞춤) |

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

for ax, s in zip(axes, ['count', 'frequency', 'probability', 'density']):
    sns.histplot(data=df, x='mpg', bins=20, stat=s, ax=ax)
    ax.set_title(f"stat='{s}'")

plt.suptitle('stat 값에 따른 Y축 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `hue` — 그룹별 색상 분리

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

sns.histplot(data=df, x='mpg', bins=20, ax=axes[0])
axes[0].set_title('hue 없음')

sns.histplot(data=df, x='mpg', bins=20, hue='origin', ax=axes[1])
axes[1].set_title("hue='origin'")

plt.tight_layout()
plt.show()
```

## 박스플롯 / 바이올린 파라미터 실험

`sns.boxplot()` / `sns.violinplot()`

### `hue` — 그룹별 색상 분리 (박스플롯)

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.boxplot(data=df, x='origin', y='mpg', ax=axes[0])
axes[0].set_title('hue 없음')

sns.boxplot(data=df, x='origin', y='mpg',
            hue='origin', palette='Set2', ax=axes[1])
axes[1].set_title("hue='origin'")
axes[1].get_legend().remove()

plt.tight_layout()
plt.show()
```

### `showfliers` — 이상치 점 표시 여부 (박스플롯)

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.boxplot(data=df, x='origin', y='horsepower',
            showfliers=True, ax=axes[0])
axes[0].set_title('showfliers=True (기본값) — 이상치 점 표시')

sns.boxplot(data=df, x='origin', y='horsepower',
            showfliers=False, ax=axes[1])
axes[1].set_title('showfliers=False — 이상치 숨김')

plt.tight_layout()
plt.show()
```

### `inner` — 바이올린 내부 표시 방식

| 값 | 설명 |
|----|------|
| `'box'` | 내부에 박스플롯 (기본값) |
| `'quart'` | 사분위수 선 표시 |
| `'point'` | 개별 데이터 포인트 |
| `None` | 내부 표시 없음 |

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 5))

for ax, inner in zip(axes, ['box', 'quart', 'point', None]):
    sns.violinplot(data=df, x='origin', y='mpg',
                   inner=inner, palette='pastel', ax=ax)
    ax.set_title(f"inner={repr(inner)}")

plt.suptitle('inner 값에 따른 바이올린 내부 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `width` — 바이올린 너비

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, w in zip(axes, [0.3, 0.8, 1.2]):
    sns.violinplot(data=df, x='origin', y='mpg',
                   width=w, palette='Set3', ax=ax)
    ax.set_title(f'width={w}')

plt.suptitle('width 값에 따른 바이올린 너비 변화', y=1.02)
plt.tight_layout()
plt.show()
```

## 산점도 파라미터 실험

`sns.scatterplot()`

### `alpha` — 점 투명도 (0: 완전 투명 → 1: 완전 불투명)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, a in zip(axes, [0.1, 0.5, 1.0]):
    sns.scatterplot(data=df, x='weight', y='mpg', alpha=a, ax=ax)
    ax.set_title(f'alpha={a}')

plt.suptitle('alpha 값에 따른 점 투명도 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `hue` + `size` — 색상과 크기로 추가 변수 표현

```python
fig, axes = plt.subplots(1, 3, figsize=(17, 5))

# 기본
sns.scatterplot(data=df, x='weight', y='mpg', alpha=0.6, ax=axes[0])
axes[0].set_title('기본 (색상/크기 없음)')

# hue 추가
sns.scatterplot(data=df, x='weight', y='mpg',
                hue='origin', alpha=0.6, ax=axes[1])
axes[1].set_title("hue='origin' 추가")

# hue + size 추가
sns.scatterplot(data=df, x='weight', y='mpg',
                hue='origin', size='cylinders', alpha=0.6, ax=axes[2])
axes[2].set_title("hue='origin' + size='cylinders'")

plt.tight_layout()
plt.show()
```

### `style` — 마커 모양으로 그룹 구분

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.scatterplot(data=penguins, x='bill_length_mm', y='body_mass_g',
                hue='species', alpha=0.7, ax=axes[0])
axes[0].set_title('hue만 사용')

sns.scatterplot(data=penguins, x='bill_length_mm', y='body_mass_g',
                hue='species', style='species', alpha=0.7, ax=axes[1])
axes[1].set_title("hue + style='species' (마커 모양도 구분)")

plt.tight_layout()
plt.show()
```

## 막대그래프 파라미터 실험

`sns.barplot()`

### `errorbar` — 오차 막대 표시 방식

| 값 | 설명 |
|----|------|
| `'ci'` | 95% 신뢰 구간 (기본값) |
| `'sd'` | 표준편차 |
| `'se'` | 표준 오차 |
| `None` | 오차 막대 없음 |

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 5))

for ax, eb in zip(axes, ['ci', 'sd', 'se', None]):
    sns.barplot(data=df, x='origin', y='mpg',
                errorbar=eb, palette='Set2', ax=ax)
    ax.set_title(f'errorbar={repr(eb)}')
    ax.set_ylim(0, 40)

plt.suptitle('errorbar 값에 따른 오차 막대 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `estimator` — 집계 함수

| 값 | 설명 |
|----|------|
| `'mean'` | 평균 (기본값) |
| `'median'` | 중앙값 |
| `np.sum` | 합계 |
| `'count'` | 개수 |

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, (est, label) in zip(axes, [('mean','mean'), ('median','median'), (np.sum,'np.sum')]):
    sns.barplot(data=df, x='origin', y='mpg',
                estimator=est, errorbar=None, palette='pastel', ax=ax)
    ax.set_title(f'estimator={label}')

plt.suptitle('estimator 값에 따른 집계 기준 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `order` — 막대 순서 지정

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 기본 순서
sns.barplot(data=df, x='origin', y='mpg',
            errorbar=None, palette='Set2', ax=axes[0])
axes[0].set_title('order 없음 (기본 순서)')

# 평균 연비 높은 순으로 정렬
order = df.groupby('origin')['mpg'].mean().sort_values(ascending=False).index
sns.barplot(data=df, x='origin', y='mpg',
            order=order, errorbar=None, palette='Set2', ax=axes[1])
axes[1].set_title('order=연비 높은 순')

plt.tight_layout()
plt.show()
```

## 라인차트 파라미터 실험

`sns.lineplot()`

### `marker` — 데이터 포인트 마커 모양

자주 쓰는 마커: `'o'`(원), `'s'`(사각), `'^'`(삼각), `'D'`(다이아몬드), `None`(없음)

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

for ax, m in zip(axes, [None, 'o', 's', '^']):
    sns.lineplot(data=df, x='model_year', y='mpg',
                 marker=m, errorbar=None, ax=ax)
    ax.set_title(f'marker={repr(m)}')

plt.suptitle('marker 값에 따른 마커 모양 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `linestyle` — 선 스타일

| 값 | 모양 |
|----|------|
| `'-'` | 실선 (기본값) |
| `'--'` | 점선 |
| `':'` | 점점선 |
| `'-.'` | 일점쇄선 |

```python
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

for ax, ls in zip(axes, ['-', '--', ':', '-.']):
    sns.lineplot(data=df, x='model_year', y='mpg',
                 linestyle=ls, linewidth=2, errorbar=None, marker='o', ax=ax)
    ax.set_title(f"linestyle='{ls}'")

plt.suptitle('linestyle 값에 따른 선 스타일 변화', y=1.02)
plt.tight_layout()
plt.show()
```

### `errorbar` — 신뢰 구간 / 표준편차 범위 표시

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for ax, eb in zip(axes, [None, 'sd', 'ci']):
    sns.lineplot(data=df, x='model_year', y='mpg',
                 errorbar=eb, marker='o', ax=ax)
    ax.set_title(f'errorbar={repr(eb)}')

plt.suptitle('errorbar 값에 따른 불확실성 범위 표시', y=1.02)
plt.tight_layout()
plt.show()
```

## 색상 팔레트 실험

`palette` 파라미터는 거의 모든 seaborn 함수에서 공통으로 사용됩니다.

```python
palettes = ['Set1', 'Set2', 'Set3', 'pastel', 'muted', 'dark', 'colorblind']

fig, axes = plt.subplots(1, len(palettes), figsize=(20, 4))

for ax, pal in zip(axes, palettes):
    sns.barplot(data=df, x='origin', y='mpg',
                errorbar=None, palette=pal, ax=ax)
    ax.set_title(f"'{pal}'")
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_ylim(0, 35)

plt.suptitle('palette 값에 따른 색상 변화', y=1.02)
plt.tight_layout()
plt.show()
```
