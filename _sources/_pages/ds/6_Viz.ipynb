| 특수 (Special) | 상관행렬, 복합 | Heatmap, Combo |

## 4-1. 시각화 구성요소

| Figure | 전체 캔버스 | `fig = plt.figure()` |
| Axes | 실제 그래프 영역 | `ax = fig.add_subplot()` |
| Title | 그래프 제목 | `ax.set_title('제목')` |
| X/Y Label | 축 이름 | `ax.set_xlabel('이름')` |
| Legend | 범례 | `ax.legend()` |
| Ticks | 축 눈금 | `ax.set_xticks([...])` |
| Grid | 격자선 | `ax.grid(True)` |

**Figure + Axes 생성 패턴:**
```python
# 단일 그래프

# 다중 그래프

```

````python
# 기본 그래프 구조 예시
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot([1, 2, 3, 4], [10, 20, 15, 25], marker='o')
ax.set_title('예시 그래프', fontsize=14)
ax.set_xlabel('X축')
ax.set_ylabel('Y축')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 4-2. 비교 시각화 (Comparison)

**그룹 간 값의 차이**를 비교할 때 사용합니다.

| 차트 | 함수 | 언제 사용? | 주요 파라미터 |
|------|------|------------|---------------|
| 막대그래프 | `sns.barplot()` | 연속형 값의 그룹 평균 비교 | `x`, `y`, `hue`, `errorbar` |
| 빈도 막대 | `sns.countplot()` | 범주별 개수 비교 | `x`, `hue`, `order` |
| 수평 막대 | `sns.barplot(orient='h')` | 카테고리 이름이 길 때 | `orient='h'` |

**barplot 주요 파라미터:**

| 파라미터 | 설명 | 예시 |
|----------|------|------|
| `x`, `y` | 축 데이터 | `x='origin'`, `y='mpg'` |
| `hue` | 색상 그룹 구분 | `hue='cylinders'` |
| `order` | 범주 순서 지정 | `order=['usa','europe','japan']` |
| `errorbar` | 오차 막대 | `errorbar='sd'` / `None` |
| `estimator` | 집계 함수 | `estimator='mean'` / `'median'` |

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# barplot: 지역별 연비 평균
# default errorbar는 95% 신뢰구간이지만, errorbar=None으로 설정하여 신뢰구간 제거
# default order는 알파벳 순이므로, order 파라미터로 원하는 순서 지정
# default estimator는 평균이므로, estimator=None으로 설정하여 평균이 아닌 원본 데이터로 그래프 그리기
sns.barplot(data=mpg, x='origin', y='mpg', ax=axes[0],
            order=['usa', 'europe', 'japan'], errorbar=None)
axes[0].set_title('지역별 평균 연비')

# countplot: 지역별 차량 수
sns.countplot(data=mpg, x='origin', ax=axes[1],
              order=['usa', 'europe', 'japan'])
axes[1].set_title('지역별 차량 수')

plt.tight_layout()
plt.show()
````

## 4-3. 분포 시각화 (Distribution)

데이터의 **퍼짐 형태**와 **분포 모양**을 파악합니다.

| 차트 | 함수 | 특징 | 언제 사용? |
|------|------|------|------------|
| 히스토그램 | `sns.histplot()` | 구간별 빈도 막대 | 연속형 분포 기본 확인 |
| KDE 밀도곡선 | `sns.kdeplot()` | 연속 확률밀도 | 분포 모양·비교 |
| 박스플롯 | `sns.boxplot()` | 사분위수·이상치 | 이상치 포함 분포 |
| 바이올린 | `sns.violinplot()` | 박스 + KDE | 분포 모양까지 파악 |
| 포인트플롯 | `sns.pointplot()` | 평균·CI 점+선 | 그룹 간 평균 비교 |

### 4-3-1. 히스토그램 / KDE

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `bins` | 구간 수 | 숫자(10, 20...) 또는 `'auto'` |
| `kde` | KDE 곡선 추가 | `True` / `False` |
| `stat` | Y축 통계 | `'count'`, `'density'`, `'probability'`, `'frequency'` |
| `hue` | 색상 그룹 | 범주형 열 이름 |
| `multiple` | 다중 분포 처리 | `'layer'`, `'stack'`, `'dodge'`, `'fill'` |

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 히스토그램 + KDE
sns.histplot(data=mpg, x='mpg', bins=20, kde=True, ax=axes[0])
axes[0].set_title('연비 분포 (Histogram + KDE)')

# 그룹별 KDE
sns.kdeplot(data=mpg, x='mpg', hue='origin', ax=axes[1])
axes[1].set_title('지역별 연비 KDE')

plt.tight_layout()
plt.show()
````

### 4-3-2. 박스플롯 / 바이올린플롯

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `x`, `y` | 범주형 / 수치형 축 | 열 이름 |
| `hue` | 색상 그룹 | 범주형 열 |
| `showfliers` | 이상치 점 표시 (boxplot) | `True` / `False` |
| `inner` | 내부 표시 방식 (violinplot) | `'box'`, `'quart'`, `'point'`, `None` |
| `width` | 박스/바이올린 너비 | 0~1 float |
| `palette` | 색상 팔레트 | `'Set1'`, `'pastel'` 등 |

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 박스플롯
sns.boxplot(data=penguins, x='species', y='body_mass_g',
            hue='sex', ax=axes[0], palette='Set2')
axes[0].set_title('종별 몸무게 분포 (Box Plot)')

# 바이올린플롯
sns.violinplot(data=penguins, x='species', y='body_mass_g',
               hue='sex', ax=axes[1], inner='quart', split=True)
axes[1].set_title('종별 몸무게 분포 (Violin Plot)')

plt.tight_layout()
plt.show()
````

### 4-3-3. 포인트플롯 (Point Plot)

그룹별 **평균**과 **신뢰구간(CI)**을 점과 선으로 표시합니다.

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `x`, `y` | 축 데이터 | 열 이름 |
| `hue` | 색상 그룹 | 범주형 열 |
| `errorbar` | 신뢰구간 표시 | `'ci'`, `'sd'`, `None` |
| `markers` | 점 모양 | `'o'`, `'s'`, `'^'` |
| `linestyles` | 선 스타일 | `'-'`, `'--'`, `':'` |

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 포인트플롯: 그룹별 평균 비교
sns.pointplot(data=mpg, x='origin', y='mpg',
              hue='cylinders', ax=axes[0], errorbar=None)
axes[0].set_title('지역·실린더별 평균 연비')

# 신뢰구간 포함
sns.pointplot(data=penguins.dropna(), x='species', y='body_mass_g',
              hue='sex', ax=axes[1], errorbar='ci')
axes[1].set_title('종·성별 평균 몸무게 (95% CI)')

plt.tight_layout()
plt.show()
````

## 4-4. 추세 시각화 (Trend / Time Series)

**시간에 따른 변화**와 **추세**를 파악합니다. 선 그래프가 핵심입니다.

| 차트 | 함수 | 특징 |
|------|------|------|
| 선 그래프 | `sns.lineplot()` | 시계열·연속 변화 기본 |
| 선 그래프 | `plt.plot()` | matplotlib 기본 |

**lineplot 주요 파라미터:**

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `x`, `y` | 축 데이터 | 열 이름 |
| `hue` | 색상 그룹 (여러 선) | 범주형 열 |
| `marker` | 데이터 포인트 표시 | `'o'`, `'s'`, `'^'` |
| `linestyle` | 선 스타일 | `'-'`, `'--'`, `':'` |
| `errorbar` | 신뢰구간 | `'ci'`, `'sd'`, `None` |
| `size` | 선 굵기 그룹 | 범주형 열 |

> **주의**: 선 그래프는 연속형 데이터에 적합. 범주형 데이터에 사용하면 오해 유발

````python
# flights 데이터: 연도별 월간 탑승객 수
flights_pivot = flights.pivot_table(
    index='month', columns='year', values='passengers'
)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 연도별 탑승객 추세
sns.lineplot(data=flights, x='year', y='passengers',
             errorbar='sd', ax=axes[0])
axes[0].set_title('연도별 평균 탑승객 수 (with SD)')

# 월별 추세 (여러 연도)
for year in [1949, 1952, 1955, 1958, 1960]:
    subset = flights[flights['year'] == year]
    axes[1].plot(subset['month'], subset['passengers'],
                marker='o', label=str(year))
axes[1].set_title('월별 탑승객 추세 (연도별)')
axes[1].legend()

plt.tight_layout()
plt.show()
````

## 4-5. 관계 시각화 (Relationship)

**변수 간 상관관계**와 **분포를 함께** 파악합니다.

| 차트 | 함수 | 특징 |
|------|------|------|
| 산점도 | `sns.scatterplot()` | 두 연속형 변수 관계 |
| 스트립플롯 | `sns.stripplot()` | 범주형 + 수치형, 모든 점 표시 |
| 스웜플롯 | `sns.swarmplot()` | 스트립과 같지만 겹침 방지 |

### 4-5-1. 산점도 (Scatter Plot)

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `x`, `y` | 두 연속형 변수 | 열 이름 |
| `hue` | 색상으로 그룹 구분 | 범주형 열 |
| `size` | 점 크기로 그룹 구분 | 범주형/수치형 열 |
| `style` | 점 모양으로 그룹 구분 | 범주형 열 |
| `alpha` | 투명도 (겹침 표시) | 0~1 float |
| `palette` | 색상 팔레트 | `'Set1'`, `'deep'` 등 |

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 기본 산점도
sns.scatterplot(data=mpg, x='horsepower', y='mpg',
                hue='origin', alpha=0.7, ax=axes[0])
axes[0].set_title('마력 vs 연비 (지역별)')

# hue + size + style 복합
sns.scatterplot(data=penguins.dropna(),
                x='bill_length_mm', y='body_mass_g',
                hue='species', size='sex', style='species',
                ax=axes[1])
axes[1].set_title('부리길이 vs 몸무게 (종·성별)')

plt.tight_layout()
plt.show()
````

### 4-5-2. Strip Plot / Swarm Plot

**범주형 + 수치형** 관계를 모든 데이터 포인트를 보여주며 시각화합니다.

| 차트 | 함수 | 특징 | 언제 사용? |
|------|------|------|------------|
| Strip Plot | `sns.stripplot()` | 점이 겹칠 수 있음 | 데이터 수가 많을 때 |
| Swarm Plot | `sns.swarmplot()` | 겹침 없이 펼쳐서 표시 | 데이터 수가 적당할 때 |

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `jitter` | 점 랜덤 분산 (strip) | `True` / `False` |
| `dodge` | hue별 분리 | `True` / `False` |
| `size` | 점 크기 | 숫자 (기본 5) |
| `alpha` | 투명도 | 0~1 float |

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Strip Plot
sns.stripplot(data=penguins.dropna(), x='species', y='body_mass_g',
              hue='sex', dodge=True, jitter=True, alpha=0.6, ax=axes[0])
axes[0].set_title('종별 몸무게 분포 (Strip Plot)')

# Swarm Plot
sns.swarmplot(data=penguins.dropna(), x='species', y='body_mass_g',
              hue='sex', dodge=True, size=4, ax=axes[1])
axes[1].set_title('종별 몸무게 분포 (Swarm Plot)')

plt.tight_layout()
plt.show()
````

## 4-6. 구성 시각화 (Composition)

**전체 대비 각 부분의 비율**을 나타냅니다.

| 차트 | 함수 | 특징 | 언제 사용? |
|------|------|------|------------|
| 누적 막대 | `sns.barplot()` + `bottom` | 각 그룹의 합 비교 | 시간별 비율 변화 |
| 파이 차트 | `plt.pie()` | 전체 100% 비율 | 간단한 비율 표시 |
| 도넛 차트 | `plt.pie()` + wedge props | 파이에 가운데 구멍 | 세련된 비율 표시 |

**누적 막대 (Stacked Bar):**
```python
# pandas plot으로 누적 막대
pivot_df.plot(kind='bar', stacked=True, ax=ax)
```

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 누적 막대
origin_cyl = mpg.groupby(['origin', 'cylinders']).size().unstack(fill_value=0)
origin_cyl.plot(kind='bar', stacked=True, ax=axes[0], colormap='Set2')
axes[0].set_title('지역별 실린더 구성 (누적 막대)')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)

# 파이 차트
origin_counts = mpg['origin'].value_counts()
axes[1].pie(origin_counts, labels=origin_counts.index,
            autopct='%1.1f%%', startangle=90)
axes[1].set_title('지역별 비율 (Pie)')

# 도넛 차트
wedge_props = dict(width=0.5)
axes[2].pie(origin_counts, labels=origin_counts.index,
            autopct='%1.1f%%', startangle=90, wedgeprops=wedge_props)
axes[2].set_title('지역별 비율 (Donut)')

plt.tight_layout()
plt.show()
````

## 4-7. 특수 시각화 (Special)

### 4-7-1. 히트맵 (Heatmap)

**행열 교차 값**을 색상으로 표현합니다. 상관행렬, 교차표에 주로 사용합니다.

| 파라미터 | 설명 | 주요 값 |
|----------|------|--------|
| `data` | 2D 데이터 (행렬) | DataFrame |
| `annot` | 셀에 값 표시 | `True` / `False` |
| `fmt` | 값 형식 | `'.2f'`, `'d'` |
| `cmap` | 색상 맵 | `'coolwarm'`, `'YlOrRd'`, `'Blues'` |
| `vmin`, `vmax` | 색상 범위 | 숫자 (-1~1 for corr) |
| `linewidths` | 셀 경계선 | 숫자 (0.5 등) |
| `square` | 정사각형 셀 | `True` / `False` |

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 상관행렬 히트맵
corr = mpg.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            vmin=-1, vmax=1, square=True, ax=axes[0])
axes[0].set_title('MPG 변수 상관행렬')

# 교차표 히트맵 (flights)
flights_pivot = flights.pivot_table(
    index='month', columns='year', values='passengers'
)
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlOrRd', ax=axes[1])
axes[1].set_title('연도×월 탑승객 수 (Heatmap)')

plt.tight_layout()
plt.show()
````

### 4-7-2. 복합 차트 (Combo Charts)

여러 차트를 **같은 Axes에 겹쳐서** 표시합니다.

| 패턴 | 설명 | 코드 패턴 |
|------|------|----------|
| Bar + Line | 막대 위에 선 추가 | `barplot()` → `lineplot()` |
| Hist + KDE | 히스토그램 + 밀도곡선 | `histplot(kde=True)` |
| Box + Strip | 박스 위에 산점도 추가 | `boxplot()` → `stripplot()` |
| Twin Axis | 다른 Y축 스케일 | `ax2 = ax.twinx()` |

```python
# Twin Axis 패턴
fig, ax1 = plt.subplots()
ax1.bar(x, y1, color='blue', alpha=0.7)
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', marker='o')
```

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box + Strip 복합
sns.boxplot(data=penguins.dropna(), x='species', y='body_mass_g',
            color='lightblue', ax=axes[0], showfliers=False)
sns.stripplot(data=penguins.dropna(), x='species', y='body_mass_g',
              color='navy', alpha=0.4, jitter=True, ax=axes[0])
axes[0].set_title('Box + Strip (종별 몸무게)')

# Histogram + KDE
sns.histplot(data=mpg, x='mpg', bins=20, kde=True,
             color='steelblue', ax=axes[1])
axes[1].set_title('Histogram + KDE (연비 분포)')

plt.tight_layout()
plt.show()
````

### 색상 팔레트 (Seaborn)

| 팔레트 | 유형 | 언제 사용? |
|--------|------|------------|
| `'Set1'`, `'Set2'`, `'tab10'` | 범주형 | 구분되는 색이 필요할 때 |
| `'Blues'`, `'Greens'` | 순차형 | 한 방향으로 크기가 변할 때 |
| `'coolwarm'`, `'RdBu'` | 발산형 | 양극 값 표현 (상관계수 등) |
| `'pastel'`, `'muted'` | 부드러운 | 배경이 복잡할 때 |

### Matplotlib vs Seaborn

| 항목 | Matplotlib | Seaborn |
|------|------------|--------|
| 코드량 | 많음 | 적음 |
| 유연성 | 높음 (세부 제어) | 보통 |
| DataFrame 연동 | 수동 | 자동 (`data=df, x=col`) |
| 통계 기능 | 없음 | 내장 (CI, KDE 등) |
| 기본 스타일 | 기본 | 세련된 기본값 |

````python
# 색상 팔레트 비교
palettes = ['Set1', 'Set2', 'tab10', 'pastel', 'muted', 'deep']
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, palette in enumerate(palettes):
    sns.barplot(data=mpg, x='origin', y='mpg',
                palette=palette, ax=axes[i], errorbar=None)
    axes[i].set_title(f'palette={palette!r}')

plt.suptitle('Seaborn 팔레트 비교', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
````
