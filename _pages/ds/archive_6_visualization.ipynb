| **관계 (Relationship)** | 두 변수가 어떤 상관이 있는지 보고 싶다 | 산점도, 스트립, 스웜 |
| **구성 (Composition)** | 전체에서 각 부분의 비율을 보고 싶다 | 파이차트, 도넛, 누적 막대 |
| **특수 (Specialized)** | 상관관계 행렬, 복합 차트 등 | 히트맵, 콤보 차트 |

## 기본 설정 및 데이터 준비

`````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (Colab에서 폰트 설치 후 주석 해제)
# plt.rc('font', family='NanumBarunGothic')

df = sns.load_dataset('mpg')
df.head()
`````

## Matplotlib 기본 구조

모든 그래프는 아래 구조로 작성합니다.

```python
fig, ax = plt.subplots(figsize=(가로, 세로))  # Figure·Axes 생성
ax.plot(...)                                   # 그래프 그리기

ax.set_xlabel('X축')
ax.set_ylabel('Y축')
plt.tight_layout()
plt.show()
```

### 자주 쓰는 공통 설정 함수

| 함수 | 역할 |
|------|------|
| `plt.figure(figsize=(w, h))` | 캔버스 크기 설정 |
| `ax.set_title('제목')` | 차트 제목 |
| `ax.set_xlabel('이름')` | X축 레이블 |
| `ax.set_ylabel('이름')` | Y축 레이블 |
| `ax.set_xlim(min, max)` | X축 범위 고정 |
| `ax.set_ylim(min, max)` | Y축 범위 고정 |
| `ax.legend()` | 범례 표시 |
| `ax.grid(True)` | 격자선 표시 |
| `plt.tight_layout()` | 여백 자동 조정 |
| `plt.savefig('파일명.png', dpi=150)` | 파일로 저장 |

## 비교 (Comparison)

값의 **크기, 순위, 차이**를 그룹 간에 비교할 때 사용합니다.

### 언제 어떤 차트를?

| 상황 | 차트 | matplotlib | seaborn |
|------|------|-----------|---------|
| 범주별 값 비교 (단일 그룹) | 막대그래프 | `ax.bar(x, height)` | `sns.barplot(data, x, y)` |
| 범주별 값 비교 (다중 그룹) | 그룹 막대그래프 | `ax.bar()` 여러 번 | `sns.barplot(hue=...)` |
| 항목이 많아 가로가 더 읽기 쉬울 때 | 수평 막대그래프 | `ax.barh(y, width)` | `sns.barplot(orient='h')` |
| 범주별 빈도수 비교 | 카운트 플롯 | `ax.bar()` + `value_counts()` | `sns.countplot(data, x)` |

### 주요 파라미터

| 파라미터 | 설명 | 예시 |
|---------|------|------|
| `x` | X축 컬럼명 | `x='origin'` |
| `y` | Y축 컬럼명 | `y='mpg'` |
| `hue` | 그룹 구분 컬럼 (색상 분리) | `hue='origin'` |
| `estimator` | 집계 함수 (기본: mean) | `estimator=np.sum` |
| `errorbar` | 오차 막대 (`'ci'`, `'sd'`, `None`) | `errorbar=None` |
| `palette` | 색상 팔레트 | `palette='Set2'` |
| `edgecolor` | 막대 테두리 색 | `edgecolor='black'` |

`````python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# 막대그래프: 원산지별 평균 연비
sns.barplot(data=df, x='origin', y='mpg', errorbar=None, palette='Set2', ax=axes[0])
axes[0].set_title('원산지별 평균 연비')
axes[0].set_xlabel('원산지')
axes[0].set_ylabel('평균 mpg')

# 그룹 막대그래프: 원산지 × 실린더 수별 (4·6실린더만)
df_sub = df[df['cylinders'].isin([4, 6])]
sns.barplot(data=df_sub, x='origin', y='mpg', hue='cylinders', errorbar=None, palette='pastel', ax=axes[1])
axes[1].set_title('원산지 × 실린더별 평균 연비')
axes[1].set_xlabel('원산지')
axes[1].set_ylabel('평균 mpg')
axes[1].legend(title='실린더')

plt.tight_layout()
plt.show()
`````

## 분포 (Distribution)

단일 변수의 **분포 형태, 중심, 퍼짐, 이상치**를 확인할 때 사용합니다.

### 언제 어떤 차트를?

| 상황 | 차트 | matplotlib | seaborn |
|------|------|-----------|---------|
| 구간별 빈도 확인 | 히스토그램 | `ax.hist(data, bins)` | `sns.histplot(data, x, bins)` |
| 부드러운 분포 곡선 | KDE 플롯 | — | `sns.kdeplot(data, x)` |
| 히스토그램 + KDE 동시 | — | — | `sns.histplot(..., kde=True)` |
| 사분위수·이상치 확인 | 박스플롯 | `ax.boxplot(data)` | `sns.boxplot(data, x, y)` |
| 분포 밀도 + 사분위수 동시 | 바이올린 플롯 | — | `sns.violinplot(data, x, y)` |

### 주요 파라미터

| 파라미터 | 설명 | 적용 함수 |
|---------|------|----------|
| `bins` | 구간 수 (기본 auto) | `hist`, `histplot` |
| `kde` | KDE 곡선 오버레이 | `sns.histplot` |
| `fill` | 곡선 아래 채우기 | `sns.kdeplot` |
| `hue` | 그룹별 분포 겹쳐 보기 | `histplot`, `kdeplot`, `boxplot`, `violinplot` |
| `inner` | 바이올린 내부 표시 (`'box'`, `'point'`, `'quart'`) | `sns.violinplot` |
| `showfliers` | 이상치 점 표시 여부 | `sns.boxplot` |

`````python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))

# 히스토그램 + KDE
sns.histplot(data=df, x='mpg', bins=20, kde=True, ax=axes[0][0])
axes[0][0].set_title('연비 분포 (히스토그램 + KDE)')

# 그룹별 KDE
sns.kdeplot(data=df, x='mpg', hue='origin', fill=True, alpha=0.4, ax=axes[0][1])
axes[0][1].set_title('원산지별 연비 분포 (KDE)')

# 박스플롯
sns.boxplot(data=df, x='origin', y='mpg', palette='Set3', ax=axes[1][0])
axes[1][0].set_title('원산지별 연비 (박스플롯)')

# 바이올린 플롯
sns.violinplot(data=df, x='origin', y='mpg', inner='box', palette='pastel', ax=axes[1][1])
axes[1][1].set_title('원산지별 연비 (바이올린 플롯)')

plt.tight_layout()
plt.show()
`````

## 추세 / 시계열 (Trend)

**시간 또는 순서**에 따른 값의 변화를 확인할 때 사용합니다.

### 언제 어떤 차트를?

| 상황 | 차트 | matplotlib | seaborn |
|------|------|-----------|---------|
| 시간에 따른 단일 변수 추세 | 라인차트 | `ax.plot(x, y)` | `sns.lineplot(data, x, y)` |
| 시간에 따른 다중 그룹 추세 | 멀티 라인 | `ax.plot()` 여러 번 | `sns.lineplot(hue=...)` |
| 추세 + 신뢰 구간 | — | — | `sns.lineplot(errorbar='ci')` |
| 면적 강조 | 영역 차트 | `ax.fill_between(x, y1, y2)` | — |

### 주요 파라미터

| 파라미터 | 설명 | 예시 |
|---------|------|------|
| `x`, `y` | 축 데이터 | `x='model_year'`, `y='mpg'` |
| `hue` | 그룹별 선 분리 | `hue='origin'` |
| `errorbar` | 신뢰 구간 (`'ci'`, `'sd'`, `None`) | `errorbar='sd'` |
| `marker` | 데이터 포인트 마커 | `marker='o'` |
| `linewidth` | 선 두께 | `linewidth=2` |
| `linestyle` | 선 스타일 (`'-'`, `'--'`, `':'`) | `linestyle='--'` |

`````python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# 연도별 평균 연비 추세
sns.lineplot(data=df, x='model_year', y='mpg', errorbar='sd', marker='o', ax=axes[0])
axes[0].set_title('연도별 평균 연비 추세')
axes[0].set_xlabel('모델 연도')
axes[0].set_ylabel('mpg')

# 원산지별 연도별 추세
sns.lineplot(data=df, x='model_year', y='mpg', hue='origin', marker='o', errorbar=None, ax=axes[1])
axes[1].set_title('원산지별 연도별 연비 추세')
axes[1].set_xlabel('모델 연도')
axes[1].set_ylabel('mpg')
axes[1].legend(title='원산지')

plt.tight_layout()
plt.show()
`````

## 관계 (Relationship)

**두 변수 간의 상관관계 또는 패턴**을 확인할 때 사용합니다.

### 언제 어떤 차트를?

| 상황 | 차트 | matplotlib | seaborn |
|------|------|-----------|---------|
| 두 수치형 변수 관계 확인 | 산점도 | `ax.scatter(x, y)` | `sns.scatterplot(data, x, y)` |
| 산점도 + 회귀선 동시 | 회귀 산점도 | — | `sns.regplot(data, x, y)` |
| 범주형 × 수치형 관계 (점) | 스트립 플롯 | — | `sns.stripplot(data, x, y)` |
| 스트립 플롯 (점 겹침 방지) | 스웜 플롯 | — | `sns.swarmplot(data, x, y)` |
| 여러 변수 쌍 관계 한 번에 | 페어플롯 | — | `sns.pairplot(df)` |

### 주요 파라미터

| 파라미터 | 설명 | 예시 |
|---------|------|------|
| `x`, `y` | 두 변수 | `x='weight'`, `y='mpg'` |
| `hue` | 색상으로 그룹 구분 | `hue='origin'` |
| `size` | 점 크기로 세 번째 변수 표현 | `size='cylinders'` |
| `alpha` | 점 투명도 (0~1) | `alpha=0.5` |
| `jitter` | 점 겹침 방지용 흔들기 | `jitter=True` |
| `vars` | pairplot에서 볼 컬럼 지정 | `vars=['mpg','weight']` |

`````python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# 산점도: 무게 vs 연비 (원산지별 색상)
sns.scatterplot(data=df, x='weight', y='mpg', hue='origin', alpha=0.6, ax=axes[0])
axes[0].set_title('차량 무게 vs 연비')
axes[0].set_xlabel('weight')
axes[0].set_ylabel('mpg')

# 스트립 플롯: 원산지별 연비
sns.stripplot(data=df, x='origin', y='mpg', hue='origin', jitter=True, alpha=0.6, ax=axes[1])
axes[1].set_title('원산지별 연비 (스트립 플롯)')
axes[1].set_xlabel('원산지')
axes[1].set_ylabel('mpg')
axes[1].get_legend().remove()

plt.tight_layout()
plt.show()
`````

## 구성 (Composition)

**전체에서 각 부분이 차지하는 비율**을 표현할 때 사용합니다.

### 언제 어떤 차트를?

| 상황 | 차트 | matplotlib | seaborn |
|------|------|-----------|---------|
| 범주별 비율 (항목 ≤5) | 파이 차트 | `ax.pie(sizes)` | — |
| 파이 차트 + 중앙 비우기 | 도넛 차트 | `ax.pie(wedgeprops={'width':0.5})` | — |
| 그룹 간 구성 비율 비교 | 누적 막대 (100%) | `df.plot(kind='bar', stacked=True)` | — |
| 시간에 따른 구성 변화 | 누적 영역 차트 | `ax.stackplot(x, y1, y2, ...)` | — |

### 주요 파라미터 (파이 차트)

| 파라미터 | 설명 | 예시 |
|---------|------|------|
| `labels` | 각 조각 레이블 | `labels=['usa','japan','europe']` |
| `autopct` | 비율 표시 형식 | `autopct='%1.1f%%'` |
| `explode` | 조각 분리 효과 | `explode=[0.05, 0, 0]` |
| `startangle` | 시작 각도 | `startangle=90` |
| `wedgeprops` | 도넛 차트용 두께 | `wedgeprops={'width': 0.5}` |

`````python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
origin_counts = df['origin'].value_counts()

# 파이 차트
axes[0].pie(
    origin_counts.values,
    labels=origin_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Set2')
)
axes[0].set_title('원산지 비율 (파이 차트)')

# 도넛 차트
axes[1].pie(
    origin_counts.values,
    labels=origin_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel'),
    wedgeprops={'width': 0.5}
)
axes[1].set_title('원산지 비율 (도넛 차트)')

plt.tight_layout()
plt.show()
`````

`````python
# 누적 막대그래프: 연도별 원산지 구성 비율
year_origin = df.groupby(['model_year', 'origin']).size().unstack(fill_value=0)
year_origin_pct = year_origin.div(year_origin.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots(figsize=(12, 5))
year_origin_pct.plot(kind='bar', stacked=True, colormap='Set2', ax=ax, width=0.8)
ax.set_title('연도별 원산지 구성 비율 (누적 막대)')
ax.set_xlabel('모델 연도')
ax.set_ylabel('비율 (%)')
ax.legend(title='원산지', bbox_to_anchor=(1.01, 1))
plt.tight_layout()
plt.show()
`````

## 특수 (Specialized)

### 히트맵 (Heatmap)

**수치형 변수 간 상관관계** 또는 **2차원 행렬 데이터**를 색상으로 표현합니다.

| 파라미터 | 설명 | 예시 |
|---------|------|------|
| `data` | 2D 데이터프레임 (상관행렬 등) | `df.corr()` |
| `annot` | 셀에 값 표시 여부 | `annot=True` |
| `fmt` | 표시 형식 | `fmt='.2f'` |
| `cmap` | 색상 팔레트 | `cmap='coolwarm'` |
| `center` | 색상 기준점 | `center=0` |
| `vmin`, `vmax` | 색상 범위 고정 | `vmin=-1, vmax=1` |
| `linewidths` | 셀 구분선 두께 | `linewidths=0.5` |
| `mask` | 특정 셀 가리기 | `mask=np.triu(np.ones_like(corr))` |

`````python
corr = df.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr, dtype=bool))  # 상삼각 마스크 (중복 제거)

fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(
    corr, mask=mask, annot=True, fmt='.2f',
    cmap='coolwarm', center=0, vmin=-1, vmax=1, linewidths=0.5, ax=ax
)
ax.set_title('수치형 변수 간 상관관계 히트맵')
plt.tight_layout()
plt.show()
`````

### 콤보 차트 (Combo Chart)

**두 가지 차트 유형을 한 그래프에** 겹쳐 표현합니다. 주로 막대 + 라인 조합에 사용합니다.

| 사용 방법 | 설명 |
|----------|------|
| `ax2 = ax.twinx()` | Y축을 공유하는 두 번째 Axes 생성 |
| 첫 번째 `ax` | 막대그래프 등 주 데이터 |
| 두 번째 `ax2` | 라인차트 등 보조 데이터 (다른 Y축 스케일) |

`````python
year_data = df.groupby('model_year').agg(
    mpg=('mpg', 'mean'),
    count=('mpg', 'size')
).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 5))

# 막대그래프: 연도별 차량 수
ax1.bar(year_data['model_year'], year_data['count'], color='steelblue', alpha=0.6, label='차량 수')
ax1.set_xlabel('모델 연도')
ax1.set_ylabel('차량 수', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')

# 라인차트: 연도별 평균 연비 (보조 Y축)
ax2 = ax1.twinx()
ax2.plot(year_data['model_year'], year_data['mpg'], color='tomato', marker='o', linewidth=2, label='평균 연비')
ax2.set_ylabel('평균 mpg', color='tomato')
ax2.tick_params(axis='y', labelcolor='tomato')

plt.title('연도별 차량 수 & 평균 연비 (콤보 차트)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.tight_layout()
plt.show()
`````

### 데이터 구성 → 차트 선택 가이드

| 변수 구성 | 목적 | 추천 차트 | 함수 |
|----------|------|----------|------|
| 범주형 1개 | 빈도 비교 | 막대그래프 | `sns.countplot(x=범주)` |
| 범주형 1개 | 비율 | 파이/도넛 | `ax.pie(sizes)` |
| 수치형 1개 | 분포 | 히스토그램 | `sns.histplot(x=수치)` |
| 수치형 1개 | 분포 + 밀도 | 히스토그램+KDE | `sns.histplot(kde=True)` |
| 수치형 1개 | 사분위수·이상치 | 박스플롯 | `sns.boxplot(y=수치)` |
| 수치형 2개 | 상관관계 | 산점도 | `sns.scatterplot(x=수치1, y=수치2)` |
| 수치형 다수 | 상관관계 행렬 | 히트맵 | `sns.heatmap(df.corr())` |
| 수치형 다수 쌍 | 쌍별 관계 | 페어플롯 | `sns.pairplot(df)` |
| 범주형 × 수치형 | 그룹별 분포 | 박스/바이올린 | `sns.boxplot(x=범주, y=수치)` |
| 범주형 × 수치형 | 그룹별 평균 비교 | 막대그래프 | `sns.barplot(x=범주, y=수치)` |
| 시간 × 수치형 | 추세 | 라인차트 | `sns.lineplot(x=시간, y=수치)` |
| 시간 × 범주 × 수치 | 그룹별 추세 | 멀티 라인 | `sns.lineplot(hue=범주)` |
| 범주 구성 변화 (시간) | 비율 변화 | 누적 막대 | `df.plot(kind='bar', stacked=True)` |

### matplotlib vs seaborn 선택 기준

| 상황 | 추천 |
|------|------|
| 세밀한 커스터마이징이 필요할 때 | matplotlib |
| 통계적 시각화 (신뢰 구간, KDE 등) | seaborn |
| 빠르게 탐색용 그래프를 그릴 때 | seaborn |
