## Solution2 

😄 설명포인트

- 딕셔너리는 key와 value가 쌍으로 매핑되어 있는 구조이다. `{key1:value1, key2:value2,...}`
- 중괄호로 묶어준다.

````python
# 방법 1 : 가장 기본
my_dict1 = {
    '이름': '파이썬',
    '멤버': ['리사','딕스','튜퍼','문엘','수형'],
    '나이': [18, 16, 18, 18, 14],
    '팬클럽': '컴기'
}
print(my_dict1)
````

````python
# 방법 2 : for loop으로 문자열 -> 정수로 변환하기
python={'이름':'파이썬', \
        '멤버': '리사, 딕스, 튜퍼, 문엘, 수형'.split(', '), \
        '나이': [int(i) for i in '18, 16, 18, 18, 14'.split(', ')], \ # list comprehension (list를 생성할 때 한 줄에 간결하게 작성하는 것)
        '팬클럽': '컴기'}
print(python)
````

````python
# 방법 3 : map함수 사용하여 한번에 문자열 -> 정수 변환
my_dict2 = {
    '이름': '파이썬',
    '멤버': '리사, 딕스, 튜퍼, 문엘, 수형'.split(', '),
    '나이': list(map(int, '18, 16, 18, 18, 14'.split(', '))), # ⚠️ 문자열을 정수로 변환하는 map함수 사용 ⚠️ map함수의 return값은 map object이므로 list로 변환하기
    '팬클럽': '컴기'
}

print(my_dict2)
````

````python
# 방법 4: zip함수 사용하기
keys = ['이름','멤버','나이','팬클럽']
values = ['파이썬', ['리사','딕스','튜퍼','문엘','수형'], [18, 16, 18, 18, 14], '컴기']
# ⚠️ 설명포인트: zip(keys, values)은 아래처럼 각 리스트에서 같은 위치의 값들을 짝지어줍니다
# zip(keys, values) → ('이름', '파이썬'), ('멤버', ['리사','딕스','튜퍼','문엘','수형']),.....
# 이렇게 짝지어진 값들을 for 문으로 하나씩 꺼내서 딕셔너리에 넣는 것입니다:
# my_dict3 = {key: value for key, value in zip(keys, values)}
my_dict3 = {}
for key, value in zip(keys, values):
  my_dict3[key] = value

print(my_dict3)
````

- 딕셔너리에는 키가 없으면 key와 value 생성(추가)이 되고, 키가 있으면 value가 업데이트 된다.
- 리스트의 count()로 해결할 수도 있음.

````python
vote=input('Input: ').split()
group_vote=dict()
for name in vote:
  if name in group_vote.keys():
    group_vote[name]+=1 # value 업데이트
  else:
    group_vote[name]=1 #key와 value 생성(추가)

print(group_vote)
````

## Problem6: 편의점 재고관리 프로그램
😄 다음 편의점 아이템을 dictionary 타입으로 만들고 sample과 같이 동작하는 프로그램을 만드시오. 메뉴번호와 판매개수는 정수로, 물건명은 문자열로 입력합니다.

```python
{"커피": 7, "펜": 3, "종이컵": 2, "우유": 1, "콜라": 5, "책": 5}
```

## Solution6

````python
store = {'커피' : 7, '펜' : 3, '종이컵' : 2, '우유' : 1, '콜라' : 5, '책' : 5}
print('-'*10, '메뉴', '-'*10)
print('1. 재고확인 2. 판매 3. 종료')
print('-'*25)

while True :
  menu = input('메뉴를 선택하세요: ')
  if menu== '3':
    print('종료합니다.')
    break
  elif menu == '1':
    product = input('물건명 입력: ')
    if product in store:
      print(f"{product}는 {store[product]}개 남았습니다.")
    else:
      print('찾는 물건이 없습니다.')
  elif menu == '2':
    product = input('물건명 입력: ')
    amount_sell = int(input(product+ '의 판매 개수: '))
    store[product] = store[product] - amount_sell
    amount_left = store[product]
    print(f"{product}는 {amount_left}개 남았습니다.")

  print()  # indentation 주의
````

- 20라인과 27라인을 설명하면서 딕셔너리를 함수호출로 사용하는 팁을 설명합니다.

````python
store = {'커피' : 7, '펜' : 3, '종이컵' : 2, '우유' : 1, '콜라' : 5, '책' : 5}
print('-'*10, '메뉴', '-'*10)
print('1. 재고확인 2. 판매 3. 종료')
print('-'*25)

def inventory():
  product = input('물건명 입력: ')
  if product in store:
    print(f"{product}는 {store[product]}개 남았습니다.")
  else:
    print('찾는 물건이 없습니다.')

def sale():
  product = input('물건명 입력: ')
  amount_sell = int(input(product+ '의 판매 개수: '))
  store[product] = store[product] - amount_sell
  amount_left = store[product]
  print(f"{product}는 {amount_left}개 남았습니다.")

a= {'1':inventory, '2':sale}  ###
while True :
  menu = input('메뉴를 선택하세요: ')
  if menu== '3':
    print('종료합니다.')
    break
  else :
   a[menu]()  ###

  print()
````

## Solution7

````python
def idol_vote():
  vote=input('Input: ').split()
  groupName=input('그룹이름: ')
  group_vote=dict()
  for name in vote:
    if name in group_vote.keys():
      group_vote[name]+=1
    else:
      group_vote[name]=1

  return group_vote, groupName

def member_top(idol, group):
  max_vote = float('-inf')
  for name in idol.keys():
    if idol[name]>max_vote:
      idol_name = name
      max_vote = idol[name]
  print(f'{group} 그룹의 인기투표 1위는 {idol_name} {max_vote}표입니다.')

idolDict, groupName = idol_vote()
member_top(idolDict,groupName)
````

## Problem8: 집합 가지고 놀기
😄 세 줄에 걸쳐 각 집합(순서대로 A, B, C)에 해당하는 원소들이 들어옵니다. 각 줄은 공백으로 구분된 1에서 100 사이의 정수 1-30개로 이루어집니다.

각 줄 안에서 원소가 겹칠 수 있는데, 이 때 그 줄에 해당하는 집합에서는 그 원소가 1개씩만 있다고 생각해야 합니다. 가령 첫 줄에 ‘3 4 5 4 3’이 들어오면 A={3, 4, 5}가 됩니다.

세 집합 A, B, C가 있을 때 A∩B∩C의 기수와 (A∪B∪C)-A의 기수를 구해 봅시다. 기수는 원소 개수를 의미합니다.

---

<font color='blue'>`test case-1`</font>

Input Sample
```
2 4 5
1 2 2 1 10 4
3 3 3 4
```

Output Sample
```
A:
B:
C:
A∩B∩C의 기수: 1
(A∪B∪C)-A의 기수: 3
```

<font color='blue'>`test case-2`</font>

Input Sample
```
20 25 29 21 16
1 4 5 6 9 25
4 7 7 9 12 21 18 15
```

Output Sample
```
A:
B:
C:
A∩B∩C의 기수: 0
(A∪B∪C)-A의 기수: 9
```


## Solution8

````python
a = set(input('A: ').split())
b = set(input('B: ').split())
c = set(input('C: ').split())
print (f'A∩B∩C의 기수: {len(a & b & c)}')
print (f'(A∪B∪C)-A의 기수: {len(a|b|c)-len(a)}')
````
