세 집합 A, B, C가 있을 때 A∩B∩C의 기수와 (A∪B∪C)-A의 기수를 구해 봅시다. 기수는 원소 개수를 의미합니다.

---

<font color='blue'>`test case-1`</font>

Input Sample
```
2 4 5
1 2 2 1 10 4
3 3 3 4
```

Output Sample
```
A:
B:
C:
A∩B∩C의 기수: 1
(A∪B∪C)-A의 기수: 3
```

<font color='blue'>`test case-2`</font>

Input Sample
```
20 25 29 21 16
1 4 5 6 9 25
4 7 7 9 12 21 18 15
```
