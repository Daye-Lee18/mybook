### Problem 1: reg_id_check()함수 만들기


😄 문자열이 주어졌을 때, 그것이 아래의 <아이디 조건>들에 모두 부합하는지 확인하는 reg_id_check()함수를 정의하시오. 입력과 출력은 함수 안에서 이루어지며, <font color='red'>결과는 True, False로 반환</font>합니다.
```
<아이디 조건>
(1) 문자열의 길이는 9-11 (inclusive).
(2) 문자열의 첫 7글자는 'student' (case sensitive).
(3) 문자열의 마지막 1글자는 '!'.
```
<font color='blue'>`test case-1`</font>

Input Sample
```
student33!
```

Output Sample
```
Input ID:
사용가능한 ID입니다.
```
<font color='blue'>`test case-2`</font>

Input Sample
```
student!
```

Output Sample
```
Input ID:
사용할 수 없는 ID입니다.
```

````python
def reg_id_check():
  user_id = input('Input ID: ')

  if len(user_id) >= 9 and len(user_id) <= 11 and user_id[:7] == 'student' and user_id[-1] == '!':
    print('사용가능한 ID입니다.')
    return True
  else:
    print('사용할 수 없는 ID입니다.')
    return False

result = reg_id_check()
# 코랩에서 return 할당 안 할 시 출력해버리는 특성 때문에 변수에 할당.
````

### Problem2: reg_pw_check()함수 만들기
😄 문자열이 주어졌을 때, 그것이 아래의 <패스워드 조건>들에 모두 부합하는지 확인하는 reg_pw_check()함수를 정의하시오. 입력과 출력은 함수 안에서 이루어지며, <font color='red'>결과는 True, False로 반환</font>합니다.

```
<패스워드 조건>
(1) 첫 글자는 반드시 대문자여야 함
(2) 영문자가 최소 1개 이상
(3) 숫자가 최소 1개 이상
(4) 특수문자 *, &, ^, % 중 하나 이상 포함
(5) 오직 영문자 + 숫자 + *, &, ^, %로만 구성되어야 함
```

```

```

````python
import string

def reg_pw_check():
    letters = string.ascii_letters  # 영문 대소문자
    digits = string.digits          # 숫자
    specials = "*&^%"
    allowed_chars = letters + digits + specials

    pw = input("Input password: ")
    if (not pw) or (not pw[0].isupper()):
      print('형식에 맞지않는 패스워드입니다.')
      return False

    has_letter = False
    has_digit = False
    has_special = False

    for char in pw:
        if char not in allowed_chars:
          print('형식에 맞지않는 패스워드입니다.')
          return False

        if char in letters:
            has_letter = True
        elif char in digits:
            has_digit = True
        elif char in specials:
            has_special = True

    if has_letter and has_digit and has_special:
      print('사용가능한 패스워드입니다.')
      return True
    else:
      print('형식에 맞지않는 패스워드입니다.')
      return False

result = reg_pw_check()
````

````python
import string

def reg_pw_check():
    password = input('Input password: ')

    all_letters = string.ascii_letters
    all_digits = string.digits
    all_special = "*&^%"

    include_one_letter = False
    include_one_digits = False
    include_one_special = False

    # 조건 (1) 첫 글자는 반드시 대문자
    if ord(password[0]) >= ord('A') and ord(password[0]) <= ord('Z'):
        include_one_letter = True

        for char in password:
            if char in all_letters:
                include_one_letter = True
            elif char in all_digits:
                include_one_digits = True
            elif char in all_special:
                include_one_special = True
            else: # 조건 (5) 오직 영문자 + 숫자 + * & ^ % 로만 구성
                print('형식에 맞지않는 패스워드입니다.')
                return False
    else:
        print('형식에 맞지않는 패스워드입니다.')
        return False


    # 조건 (2), (3), (4)
    if include_one_letter & include_one_digits & include_one_special:
        print('사용가능한 패스워드입니다.')
        return True
    else:
        print('형식에 맞지않는 패스워드입니다.')
        return False


result = reg_pw_check()
````

### Problem6 : mySum()함수 만들기

😄 파이썬에는 sum()함수가 있습니다.  
sum()함수는 sum(range(start, end)), sum([10,20,30])과 같이 범위 또는 반복가능한 숫자형데이터)를 넣어주면 합을 반환해줍니다. sum() 함수와 똑같이 동작하는 mySum()함수를 정의하시오. 입력은 정수 2개가 콤마로 구분하여 들어옵니다. 첫 번째 숫자는 range() 함수의 start값으로, 두 번째 숫자는 end값으로 사용됩니다.
출력은 start부터 end까지 수열의 합을 구합니다.
입력과 출력은 함수 밖에서 이루어집니다.

---

<font color='blue'>`test case-1`</font>

Input Sample
```
1,100
```
Output Sample
```
Input:
Output: 5050
```

<font color='blue'>`test case-2`</font>

Input Sample
```
50,100
```
Output Sample
```
Input:
Output: 3825
```

- 아래 sum()함수는 파이썬의 내장함수입니다.

````python
start=1
end=100
print(sum(range(start, end)))
print(sum([10,20,30]))
````

````python
def mySum(data):
  result = 0
  for num in data:
    result += num
  return result

start, end = map(int,input('Input: ').split(','))
data = range(start,end+1)
print(f'Output: {mySum(data)}')
````

- mySum() 함수를 정의했다면 아래 코드와 같이 전달인자로 리스트 [10,20,30,40]를 넣어서 잘 동작하는지 확인해보시오.

### Problem7: myMax() 함수 만들기
😄 파이썬에서는 max()함수가 있습니다. max()함수는  max(data)과 같이 data를 넣어주면 data에서 최대값을 반환해줍니다. max() 함수와 똑같이 동작하는 myMax()함수를 정의하시오.

---
<font color='blue'>`test case-1`</font>

Input Sample
```
80, 33, 95, 46, 89, 95, 84, 68, 4, 60
```
Output Sample
```
Input:
Max: 95
```
<font color='blue'>`test case-2`</font>

Input Sample
```
31, 39, 14, 93, 51, 62, 20, 12, 9, 3
```
Output Sample
```
Input:
Max: 93
```


- 아래 max()함수는 파이썬의 내장함수입니다.

````python
print(max(31, 39, 14, 93, 51, 62, 20, 12, 9, 3))
print(max([1,40,23,12]))
````

````python
def myMax(data):
  result = data[0]
  for num in data:
    if num > result:
      result = num
  return result

data = list(map(int, input('Input: ').split(',')))
print(f'Max: {myMax(data)}')
````
