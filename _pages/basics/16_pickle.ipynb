# Chapter 16. Pickle 

pickle은 파이썬 "객체"를 바이트 스트림(binary)으로 직렬화하는 표준 모듈이다. 

* 파이썬의 리스트, 딕셔너리(dictionary), 사용자 정의 "객체"를 파이썬이 나중에 다시 복원할 수 있는 형태로 저장하는 방식이다. 
* json과 달리 pickle은 텍스트가 아니라 바이너리 포맷이라 사람이 읽기 쉽지 않다. 
* Python 문서도 pickle을 Python object serialization 모듈이라고 설명한다. 
* pickle 문서 저장 = serialization (object -> binary)
* pickle 문서 로드 = deserialization (binary -> object)

장점 
* 여러 파이썬 객체를 저장/복원할 수 있다. 

단점 
* 보안상 위험하다. 신뢰할 수 있는 문서만 열어야한다. 

```python
import pickle 

data = {
    "name": "Jessi",
    "score": [95, 88, 100]
}

# 저장 
# pickle은 binary 포맷이라 binary mode로 연다. 
with open("data.pkl", "wb") as f:
    pickle.dump(data, f)

# 불러오기 
with open("data.pkl", "rb") as f:
    loaded = pickle.load(f)

print(loaded)
```
