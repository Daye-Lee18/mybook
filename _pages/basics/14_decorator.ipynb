## Chapter 14. Decorator 

## Decorator란?

decorator는 함수 wrapper로 “함수를 감싸서(wrapper) 실행 흐름을 바꾸는 것”이다. 실제 작동 원리는 wrapper가 argument 받는다

## Decorator 정의

### 함수를 이용한 decorator 정의

````python
def deco(func):
    def wrapper(*args, **kwargs):
        print("before")
        result = func(*args, **kwargs)
        print("after")
        return result
    return wrapper

@deco
def hello(name):
    print(f"hello {name}")
````

위의 코드에서 hello()를 실행하면 실제로는 다음 코드가 실행된다. 

```python
hello = deco(hello)
```

### 클래스를 이용한 decorator 정의

## Decorator 사용

````python
import datetime 

class DatetimeDecorator:
    def __init__(self, f):
        self.func = f 

    def __call__(self, *args, **kwargs):
        print(datetime.datetime.now())
        self.func(*args, **kwargs)
        print(datetime.datetime.now())
````

````python
# 사용 
class MainClass:
    @DatetimeDecorator
    def main_function1():
        print("main function1 start")
    
    @DatetimeDecorator
    def main_function2():
        print("main function2 start")

my = MainClass()
my.main_function1()
my.main_function2()
````

## Decorator 사용 사례 

### Flask 웹 프레임워크 라우팅 

python에서는 다양한 decorator가 제공된다. Flask 웹 프레임워크에서는 라우팅을 위한 `@app.route`라는 decorator를 사용한다. 

- `@app.route`는 URL 경로를 함수에 연결한다. Flask 어플리케이션은 특정 URL에서 요청이 들어올때, 해당 URL에 연결된 함수를 실행할 수 있다. 

````python
from flask import Flask 

app = Flask(__name__)

@app.route("/")
def hello():
    return "Hello World!"
````
